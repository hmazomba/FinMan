```
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

```
df_twtr = yf.download('TWTR',
                      start='2018-01-01',
                      end='2018-12-31',
                      progress=False,
                      auto_adjust=True)
```

```
trace1 = {
    'x': df_twtr.index,
    'open': df_twtr.Open,
    'close': df_twtr.Close,
    'high': df_twtr.High,
    'low': df_twtr.Low,
    'type': 'candlestick',
    'name': 'TWTR',
    'showlegend': True
}
```

```
avg_30 = df_twtr.Close.rolling(window=30, min_periods=1).mean()
avg_50 = df_twtr.Close.rolling(window=50, min_periods=1).mean()
ema_30 = df_twtr['Close'].ewm(span=30).mean()
ema_50 = df_twtr['Close'].ewm(span=50).mean()
```

```
trace2 = {
    'x': df_twtr.index,
    'y': avg_30,
    'type': 'scatter',
    'mode': 'lines',
    'line':{
        'width': 1,
        'color':'blue'
    },
    'name': 'Moving Average of 30 Periods'
}
```

```
trace3 = {
    'x': df_twtr.index,
    'y': avg_50,
    'type': 'scatter',
    'mode': 'lines',
    'line':{
        'width': 1,
        'color': 'red'
    },
    'name': 'Moving Average of 50 Periods'
}
```

```
trace4 = {
    'x': df_twtr.index,
    'y': ema_30,
    'type': 'scatter',
    'mode': 'lines',
    'line':{
        'width': 1,
        'color': 'green'
    },
    'name': 'Exponential Moving Average of 30 Periods'
}
```

```
trace5 = {
    'x': df_twtr.index,
    'y': ema_50,
    'type': 'scatter',
    'mode': 'lines',
    'line':{
        'width': 1,
        'color': 'yellow'
    },
    'name': 'Exponential Moving Average of 50 Periods'
}
```

```
data =[trace1, trace2, trace3, trace4, trace5]
layout = go.Layout({
    'title': {
        'text': 'TWTR - Twitter',
        'font': {
            'size': 15
        }
    }
})
```

```
fig = go.Figure(data=data, layout=layout)
fig.write_html("./TWTR-web.html")
fig.show()
```
